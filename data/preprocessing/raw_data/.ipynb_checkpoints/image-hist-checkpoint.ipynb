```python
import os
import numpy as np
import tensorflow as tf
import random as python_random

# os.environ['PYTHONHASHSEED']=str(1)
# tf.random.set_seed(1)
# np.random.seed(1)
# python_random.seed(1)

def reset_random_seeds():
   os.environ['PYTHONHASHSEED']=str(1)
   tf.random.set_seed(1)
   np.random.seed(1)
   python_random.seed(1)
```

```python
import pandas as pd
import sys
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.svm import SVC

import keras
from keras import regularizers, optimizers
from keras.layers import Lambda, Input, Dense, BatchNormalization as BN, concatenate, Dropout
from keras.models import Model
from sklearn.utils.class_weight import compute_class_weight

from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import roc_curve
import scikitplot as skplt
import matplotlib.pyplot as plt

from mlxtend.evaluate import confusion_matrix
from mlxtend.plotting import plot_confusion_matrix
from keras.models import load_model
```

```python
raw_data = pd.read_csv('MBdata_33CLINwMiss_1KfGE_1KfCNA.csv')
```

```python
def build_model_images():
    reset_random_seeds()
    
    inp = Input(shape=(int_size,))

    x = Dense(ds, activation=act)(inp)
    x = BN()(x)

    x = Dense(ds // 2, activation=act)(x)
    x = BN()(x)

    out = Dense(out_size, activation='softmax')(x)

    model = Model(inp, out, name='dnn_ge')
    model.summary()

    sgd = optimizers.SGD(lr=0.01 ,nesterov=False)
    model.compile(optimizer=sgd, loss='categorical_crossentropy', metrics=['accuracy'])
    
    return model
```

```python
batch_size = 32
act = "tanh"
ds = 32
int_size = 368
out_size = 6
epochs = 100

model= build_model_images()
model.save('model.h5')
```

```python
def model_scores(model_path, X_train, y_train, X_test, y_test):
    
    class_weights = compute_class_weight('balanced', np.unique(y_train), y_train)
    
    y_train_enc = keras.utils.to_categorical(y_train, 6)
    y_test_enc = keras.utils.to_categorical(y_test, 6)
    
    dnn=load_model(model_path)
#     dnn=model
    
    dnn.fit(X_train, y_train_enc, epochs=epochs, batch_size=batch_size, class_weight=class_weights)
    
    y_pred_prob = dnn.predict(X_test)  
    y_pred_class = y_pred_prob.argmax(axis=-1)
#     print(confusion_matrix(y_test,y_pred_class))
    cm = confusion_matrix(y_test,y_pred_class)
    fig, ax = plot_confusion_matrix(conf_mat=cm, show_absolute=False,show_normed=True)
    plt.show()
    
    train_scores = dnn.evaluate(X_train, y_train_enc)
    test_scores = dnn.evaluate(X_test, y_test_enc)
    
    print(train_scores)
    print(test_scores)

    return(train_scores, test_scores)
```

```python
def cross_val(X, y, model_path):
    train_scores = []
    test_scores = []
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        train_score, test_score = model_scores(model_path, X_train, y_train, X_test, y_test)
        train_scores.append(train_score[1])
        test_scores.append(test_score[1])   
    print(np.mean(train_scores))
    print(np.mean(test_scores))
```

```python
scaler = MinMaxScaler() 
```

# vector_k5 

```python
vector_k5 = pd.read_csv("vector_measures_k5_all_v5.csv")
```

```python
vector_k5_ND = vector_k5.drop_duplicates(subset=['catalog_ID'])
```

```python
vector_k5_ND.rename(columns={"catalog_ID": "METABRIC_ID"}, inplace=True)
```

```python
vector_k5_ND_raw_data= pd.merge(vector_k5_ND, raw_data, on='METABRIC_ID')
```

```python
vector_k5_ND_raw_data_multi= vector_k5_ND_raw_data[
    (vector_k5_ND_raw_data['Histological_Type'] == 'IDC') | 
    (vector_k5_ND_raw_data['Histological_Type'] == 'ILC') |
    (vector_k5_ND_raw_data['Histological_Type'] == 'IDC+ILC') |
    (vector_k5_ND_raw_data['Histological_Type'] == 'IDC-TUB') |
    (vector_k5_ND_raw_data['Histological_Type'] == 'IDC-MUC') |
    (vector_k5_ND_raw_data['Histological_Type'] == 'IDC-MED')]

vector_k5_ND_raw_data_multi["Histological_Type"].value_counts()
```

```python
vector_k5_ND_raw_data_multi['Histological_Type'].replace({"IDC":0,"ILC":1, "IDC+ILC":2, "IDC-MUC":3, 
                                                             "IDC-TUB":4, "IDC-MED":5}, inplace=True)
```

```python
vector_k5_ND_raw_data_multi.iloc[:,2:370] = scaler.fit_transform(vector_k5_ND_raw_data_multi.iloc[:,2:370])
```

```python
vector_k5_input = vector_k5_ND_raw_data_multi.iloc[:,2:370].values
vector_k5_label = vector_k5_ND_raw_data_multi['Histological_Type'].values
```

```python
cross_val(vector_k5_input, vector_k5_label, 'model.h5')
# cross_val(vector_k5_input, vector_k5_label, model)
```
